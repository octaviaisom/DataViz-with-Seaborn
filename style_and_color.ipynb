```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

# 1. Setting Styles

```python
sns.set_style('ticks') #white, darkgrid, whitegrid
sns.countplot(x='sex', data=tips)
sns.despine() #by default, removes top and right
```

# 2. Size and Aspect

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('poster', font_scale = 2) #2 times the size
sns.countplot(x='sex', data=tips)
```

```python
sns.set_context('notebook') #set back to normal
```

# 3. Palettes and Colors
<https://matplotlib.org/tutorials/colors/colormaps.html?highlight=colormaps>

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', palette='prism')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', palette='hsv')
```
