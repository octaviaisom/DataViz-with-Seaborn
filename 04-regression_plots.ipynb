```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':100}) #passing in matplotlib params
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex', row='time') #alternative grid call
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day', hue='sex',
          aspect=0.6, size=8) #adding size/aspect params
```
